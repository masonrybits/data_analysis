# Analysis on relationship between chance of admit and GRE score & GPA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
admission = pd.read_csv("./Admission_Predict_Ver1.1.csv")
print(f"Data Shape: {admission.shape}")
print(admission.head())
print(admission.info())
```

```python
sns.set()
plt.figure(figsize=(10,6),dpi=300)
sns.regplot(x='GRE Score', y='Chance of Admit ', data=admission, order=1)
```

```python
sns.set()
plt.figure(figsize=(10,6),dpi=300)
sns.regplot(x='CGPA', y='Chance of Admit ', data=admission, order=1)
```

## Conslusions

* Both GRE score and GPA are positively correlated to the the chance of admit.
* Looks to me that GPA is better correlated to the chace of admit than GRE score, which shows that eveything being the same, one probably wants to improve GPA to have a better chance to get into grad school.
